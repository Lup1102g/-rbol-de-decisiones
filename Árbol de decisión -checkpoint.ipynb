```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
alumnos = pd.read_csv("studen.csv")
                  
regulares = alumnos[alumnos["Promedio"]==0]                     
riesgo = alumnos[alumnos["Promedio"]==1]


plt.figure(figsize=(6, 6))
plt.xlabel('Primer', fontsize = 20.0)
plt.ylabel('Segundo', fontsize = 20.0)
plt.scatter(regulares["Primer_Parcial"], regulares["Segundo_Parcial"], 
            label="regular (Clase: 0)", marker="*", c="skyblue", s=200)
plt.scatter(riesgo["Primer_Parcial"], riesgo["Segundo_Parcial"],
            label="riesgo (Clase: 1)", marker="*", c="lightcoral", s=200)
plt.legend(bbox_to_anchor=(1, 0.15))
plt.show()
```

```python
from scipy.stats import entropy
from math import log

primerparcial = pd.Series([4, 8, 16, 19])
segundoparcial = pd.Series([4, 8, 16, 19])

print(primerparcial.value_counts()/primerparcial.size)
print(segundoparcial.value_counts()/segundoparcial.size)
print(entropy(segundoparcial.value_counts()/segundoparcial.size, base=2))
print(entropy(segundoparcial.value_counts()/segundoparcial.size, base=2))
```

```python
from sklearn.model_selection import train_test_split

datos_entrena, datos_prueba, clase_entrena, clase_prueba = train_test_split(
    alumnos[["Edad", "Ausencias"]],
    alumnos["Promedio"], 
    test_size=0.30)
```

```python
from sklearn import tree

arbol_decision = tree.DecisionTreeClassifier(criterion="entropy")

arbol = arbol_decision.fit(datos_entrena, clase_entrena)

accuracy = arbol_decision.score(datos_prueba, clase_prueba)

print(accuracy)

print(tree.export_text(arbol,
                      feature_names=["Primer_Parcial", "Segundo_Parcial"]))
plt.figure(figsize=(12, 6))
tree.plot_tree(arbol, 
              feature_names=["Primer_Parcial", "Segundo_Parcial"])
plt.show()
```
